Провести кореляційний аналіз даних, шукаючи відповіді на питання

a.	Чи справді різні фактори прогнозування, які були обрані спочатку, впливають на очікувану тривалість життя?
    Які передбачувані змінні насправді впливають на тривалість життя?

Для проведення кореляційного аналізу та відповіді на питання щодо впливу різних факторів на очікувану тривалість життя, ми можемо розглянути матрицю кореляції та зосередитися на коефіцієнтах кореляції між факторами та тривалістю життя.

```python
# Обчислення матриці кореляції між факторами та тривалістю життя
correlation_matrix = numeric_columns.corr()

# Виведення матриці кореляції
print("Матриця кореляції:")
print(correlation_matrix)

# Виведення теплокарти кореляції
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Теплокарта кореляції")
plt.show()
```

Матриця кореляції вказує на ступінь лінійної залежності між різними факторами та тривалістю життя. Значення, близькі до 1 або -1, вказують на сильну кореляцію, тоді як значення близькі до 0 свідчать про відсутність кореляції.

Виведемо коефіцієнти кореляції, які перевищують певний поріг, наприклад, 0.3 (якщо вони позитивні) або -0.3 (якщо вони негативні):

```python
# Виведення коефіцієнтів кореляції з порогом
correlation_threshold = 0.3
life_expectancy_column = 'Life expectancy '
high_correlation = (correlation_matrix[life_expectancy_column] > correlation_threshold) | (correlation_matrix[life_expectancy_column] < -correlation_threshold)

print("\nФактори, які мають високу кореляцію з тривалістю життя:")
print(correlation_matrix[life_expectancy_column][high_correlation])
```

b.	Чи повинна країна з нижчим значенням очікуваної тривалості життя (<65) збільшити витрати на охорону здоров’я, щоб покращити середню тривалість життя?

Для визначення впливу витрат на охорону здоров'я на середню тривалість життя країни з низькою очікуваною тривалістю життя (<65 років), можна розглянути середні значення витрат на охорону здоров'я для країн з низькою та високою тривалістю життя. Далі можна порівняти ці значення і визначити, чи існує статистично значуща різниця.

```python
# Вибір даних для країн з низькою очікуваною тривалістю життя
low_life_expectancy = df[df['Life expectancy '] < 65]

# Виведення кореляції між очікуваною тривалістю життя та витратами на охорону здоров'я для цієї підгрупи
correlation_low_life_health = low_life_expectancy['Life expectancy '].corr(low_life_expectancy['Total expenditure'])

print(f"Кореляція між очікуваною тривалістю життя та витратами на охорону здоров'я для країн з низькою очікуваною тривалістю життя: {correlation_low_life_health:.2f}")
```

Кореляція між очікуваною тривалістю життя та витратами на охорону здоров'я для країн з низькою очікуваною тривалістю життя складає -0.16. Це вказує на слабку негативну кореляцію між цими двома змінними. Іншими словами, зі збільшенням витрат на охорону здоров'я у країнах з низькою очікуваною тривалістю життя може спостерігатися невелике зменшення очікуваної тривалості життя, але ця залежність є досить слабкою.

c.	Як рівень дитячої та дорослої смертності впливає на тривалість життя?

Для визначення впливу рівнів дитячої та дорослої смертності на тривалість життя можна скористатися аналізом кореляції.

```python
# Виведення кореляції між тривалістю життя та рівнем дитячої та дорослої смертності
correlation_life_adult_child_mortality = df['Life expectancy '].corr(df['Adult Mortality'])
correlation_life_infant_mortality = df['Life expectancy '].corr(df['infant deaths'])

print(f"Кореляція між тривалістю життя та дорослою смертністю: {correlation_life_adult_child_mortality:.2f}")
print(f"Кореляція між тривалістю життя та смертністю немовлят: {correlation_life_infant_mortality:.2f}")
```

З огляду на отримані результати, можна стверджувати, що рівень дорослої смертності має сильний вплив на тривалість життя, тоді як вплив смертності немовлят є менш вираженим. Зменшення рівня дорослої смертності може сприяти підвищенню тривалості життя у розглянутому контексті даних.

d.	Чи має очікувана тривалість життя позитивну чи негативну кореляцію зі звичками в їжі, способом життя, фізичними вправами, курінням, вживанням алкоголю тощо.

Визначимо кореляцію між очікуваною тривалістю життя та іншими факторами, такими як звички в їжі, спосіб життя, фізичні вправи, куріння, вживання алкоголю і т.д

```python
# Вибір факторів для аналізу кореляції
selected_factors = ['Life expectancy ', ' BMI ', 'Alcohol']

# Створення підмножини даних з вибраними факторами
selected_data = df[['Life expectancy ', ' BMI ', 'Alcohol']]

# Виведення кореляційної матриці
correlation_matrix_selected = selected_data.corr()
print("Кореляційна матриця:")
print(correlation_matrix_selected)

# Виведення коефіцієнтів кореляції між тривалістю життя та вибраними факторами
correlation_life_selected_factors = correlation_matrix_selected['Life expectancy ']
print("\nКореляція між тривалістю життя та вибраними факторами:")
print(correlation_life_selected_factors)
```

Також виявлений позитивний взаємозв'язок між вживанням алкоголю та тривалістю життя, що може вказувати на можливий вплив культурних та соціальних аспектів споживання алкоголю на тривалість життя у різних країнах.

h.	Який вплив охоплення імунізацією на очікувану тривалість життя?

```python
# Визначення колонок для аналізу
health_related_columns = [
    'Life expectancy ',
    'Polio', 
    'Diphtheria ',
    'Hepatitis B',
    ' HIV/AIDS'
]

# Створення підмножини даних з вибраними факторами
selected_data_health = df[health_related_columns]

# Виведення кореляційної матриці
correlation_matrix_health = selected_data_health.corr()

# Виведення кореляції між очікуваною тривалістю життя та імунізацією/здоров'ям
correlation_life_health = correlation_matrix_health['Life expectancy ']

print("Кореляція між очікуваною тривалістю життя та імунізацією/здоров'ям:")
print(correlation_life_health)
```

Отримані результати свідчать про позитивний вплив імунізації на тривалість життя, зокрема відзначаються позитивні кореляції з вакцинацією від поліомієліту, дифтерії та гепатиту B. З іншого боку, виявлений негативний зв'язок із поширенням ВІЛ/СНІДу, що свідчить про важливість боротьби з інфекційними захворюваннями для поліпшення тривалості життя.
